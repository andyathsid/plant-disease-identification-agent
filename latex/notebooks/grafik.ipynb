```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs

# 1. Membuat Data Dummy (Simulasi Vektor 3D)
n_samples = 600
centers = 4
X, y = make_blobs(n_samples=n_samples, centers=centers, n_features=3, random_state=42, cluster_std=6.0)

# Mapping label angka ke nama tanaman
label_map = {0: 'Tomat', 1: 'Apel', 2: 'Jagung', 3: 'Anggur'}
labels = [label_map[i] for i in y]

# Buat DataFrame
df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], z=X[:,2], label=labels))

# Definisi Warna
color_map = {'Tomat': '#e74c3c', 'Apel': '#2ecc71', 'Jagung': '#f1c40f', 'Anggur': '#9b59b6'}

# ---------------------------------------------------------
# GAMBAR 1: Global Vector Space (3D) - Semua Data
# ---------------------------------------------------------
fig1 = px.scatter_3d(
    df, x='x', y='y', z='z', color='label',
    color_discrete_map=color_map,
    opacity=0.8,
    labels={'label': 'Plant Name'}
)

fig1.update_layout(
    margin=dict(l=0, r=0, b=0, t=0),
    scene=dict(
        xaxis=dict(showgrid=True, showticklabels=False, title=''),
        yaxis=dict(showgrid=True, showticklabels=False, title=''),
        zaxis=dict(showgrid=True, showticklabels=False, title='')
    ),
    legend=dict(x=0, y=1)
)

fig1.show()

# ---------------------------------------------------------
# GAMBAR 2: Filtered View (3D) - Tomat Merah, Lainnya Abu Tua
# ---------------------------------------------------------
match_df = df[df['label'] == 'Tomat']
ignored_df = df[df['label'] != 'Tomat']

fig2 = go.Figure()

# Layer 1: Data yang Terfilter (Abu-abu Tua)
fig2.add_trace(go.Scatter3d(
    x=ignored_df['x'], y=ignored_df['y'], z=ignored_df['z'],
    mode='markers',
    marker=dict(
        size=3,
        color='#4b5563', # Abu-abu tua
        opacity=0.15     # Dibuat samar tapi tetap ada strukturnya
    ),
    name='Lainnya'
))

# Layer 2: Data Tomat (Aktif/Highlight)
fig2.add_trace(go.Scatter3d(
    x=match_df['x'], y=match_df['y'], z=match_df['z'],
    mode='markers',
    marker=dict(
        size=5,
        color='#e74c3c', # Merah Tomat
        opacity=1.0,
        line=dict(width=1, color='white')
    ),
    name='Tomat'
))

fig2.update_layout(
    margin=dict(l=0, r=0, b=0, t=0),
    scene=dict(
        xaxis=dict(showgrid=True, showticklabels=False, title=''),
        yaxis=dict(showgrid=True, showticklabels=False, title=''),
        zaxis=dict(showgrid=True, showticklabels=False, title='')
    ),
    legend=dict(x=0, y=1)
)

fig2.show()
```

```python
# ---------------------------------------------------------
# GAMBAR 2: Filtered View (3D) - Hanya Tomat yang Aktif
# ---------------------------------------------------------
# Pisahkan data
match_df = df[df['label'] == 'Tomat']
ignored_df = df[df['label'] != 'Tomat']

fig2 = go.Figure()

# Layer 1: Data yang Diabaikan (Transparan/Ghost)
fig2.add_trace(go.Scatter3d(
    x=ignored_df['x'], y=ignored_df['y'], z=ignored_df['z'],
    mode='markers',
    marker=dict(
        size=4,
        color='#bdc3c7', # Abu-abu
        opacity=0.1      # Sangat transparan (efek ghost)
    ),
    name='Lainnya (Filtered Out)'
))

# Layer 2: Data Tomat (Solid & Menonjol)
fig2.add_trace(go.Scatter3d(
    x=match_df['x'], y=match_df['y'], z=match_df['z'],
    mode='markers',
    marker=dict(
        size=6,
        color='#e74c3c', # Merah Tomat
        opacity=1.0,     # Solid
        line=dict(width=2, color='white') # Outline putih biar pop-up
    ),
    name='Tomat (Active Scope)'
))

# Pengaturan Layout
fig2.update_layout(
    margin=dict(l=0, r=0, b=0, t=0),
    scene=dict(
        xaxis_title='Dimensi Vektor 1',
        yaxis_title='Dimensi Vektor 2',
        zaxis_title='Dimensi Vektor 3'
    ),
    legend=dict(x=0, y=1)
)

fig2.show()
```
