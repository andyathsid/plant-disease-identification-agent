# YOLOv11 Model Sanity Check

This notebook performs inference using the trained YOLOv11 model to verify it's working correctly.

```python
from ultralytics import YOLO
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
```

## Load Best Model Weights

Loading the best performing YOLOv11m model from training.

```python
# Load the best trained model
model_path = r"D:\Workspace\Repository\thesis\research\object-detection-engine\models\yolov11\runs\detect\train_yolov11m\weights\best.pt"
model = YOLO(model_path)

print(f"Model loaded from: {model_path}")
print(f"Model type: {model.model_name}")
```

## Download Test Image

Downloading the grape leaf image from the provided URL.

```python
# Download image from URL
image_url = "https://thesis-assets.andyathsid.com/plantwild/gallery/images/grape%20leaf/google_0246.jpg"

response = requests.get(image_url)
img = Image.open(BytesIO(response.content))

print(f"Image downloaded successfully")
print(f"Image size: {img.size}")
print(f"Image mode: {img.mode}")

# Display the image
plt.figure(figsize=(10, 8))
plt.imshow(img)
plt.axis('off')
plt.title('Original Grape Leaf Image')
plt.show()
```

## Run Inference

Performing object detection on the grape leaf image.

```python
# Run inference
results = model(img, conf=0.25, iou=0.45)

# Get the first result (since we only passed one image)
result = results[0]

print(f"\nDetection Results:")
print(f"Number of detections: {len(result.boxes)}")
print(f"\nDetected objects:")

# Print details of each detection
for i, box in enumerate(result.boxes):
    cls_id = int(box.cls[0])
    conf = float(box.conf[0])
    cls_name = result.names[cls_id]
    coords = box.xyxy[0].tolist()
    
    print(f"\nDetection {i+1}:")
    print(f"  Class: {cls_name}")
    print(f"  Confidence: {conf:.4f}")
    print(f"  Bounding box: [{coords[0]:.1f}, {coords[1]:.1f}, {coords[2]:.1f}, {coords[3]:.1f}]")
```

## Visualize Results

Display the image with bounding boxes and labels.

```python
# Plot the results using ultralytics built-in plotting
import numpy as np

# Get annotated image
annotated_img = result.plot()

# Convert from BGR to RGB (OpenCV to matplotlib format)
annotated_img_rgb = annotated_img[..., ::-1]

# Display
plt.figure(figsize=(12, 10))
plt.imshow(annotated_img_rgb)
plt.axis('off')
plt.title(f'YOLOv11m Detection Results - {len(result.boxes)} objects detected')
plt.tight_layout()
plt.show()
```

## Model Information

Display model metadata and class names.

```python
# Display model class names
print("Model Class Names:")
print(f"Total classes: {len(result.names)}")
print("\nClass mapping:")
for cls_id, cls_name in result.names.items():
    print(f"  {cls_id}: {cls_name}")
```
