# Benchmark Results Analysis

This notebook loads and displays the benchmark results for YOLOv11 and OWLv2 models.

```python
import pandas as pd
from pathlib import Path
import glob
import os

# Define the root directory where CSVs are stored (parent of benchmarking folder)
ROOT_DIR = Path("..")

# Function to find the latest CSV file matching a pattern
def find_latest_csv(pattern):
    files = list(ROOT_DIR.glob(pattern))
    if not files:
        return None
    # Sort by modification time, newest first
    latest_file = max(files, key=os.path.getmtime)
    return latest_file

# Find latest benchmark files
yolo_csv = find_latest_csv("onnx_benchmark_results_*.csv")
owlv2_csv = find_latest_csv("owlv2_benchmark_results_*.csv")

dfs = []

if yolo_csv:
    print(f"Loading YOLO results from: {yolo_csv.name}")
    df_yolo = pd.read_csv(yolo_csv)
    dfs.append(df_yolo)
else:
    print("No YOLO benchmark results found.")

if owlv2_csv:
    print(f"Loading OWLv2 results from: {owlv2_csv.name}")
    df_owl = pd.read_csv(owlv2_csv)
    dfs.append(df_owl)
else:
    print("No OWLv2 benchmark results found.")

if dfs:
    # Combine all results
    df_all = pd.concat(dfs, ignore_index=True)
    
    # Reorder columns for better readability if needed
    cols = ["Model", "Input_Size", "Phase", "Mean", "Median", "P90", "P95", "Min", "Max"]
    df_all = df_all[cols]
    
    # Display the dataframe
    display(df_all)
else:
    print("No benchmark data available.")
```

```python
# Filter for Total time to compare models
if dfs:
    df_total = df_all[df_all["Phase"] == "T_total"].sort_values("Mean")
    print("\nComparison of Total Inference Time (ms):")
    display(df_total)
```
