```python
import pandas as pd
from pathlib import Path
corpus_path = Path('data/corpus.parquet')
qa_path = Path('data/qa.parquet')
print('Corpus exists:', corpus_path.exists())
print('QA exists:', qa_path.exists())
if corpus_path.exists():
    df = pd.read_parquet(corpus_path)
    print('Corpus shape', df.shape)
    print(df.columns.tolist())
    print(df.head(2))
if qa_path.exists():
    qa = pd.read_parquet(qa_path)
    print('QA shape', qa.shape)
    print(qa.columns.tolist())
    print(qa.head(2))
```

```python
import pandas as pd

# Paths to the JSONL files
gardenology_path = 'data/gardenology.jsonl'
plantvillage_path = 'data/plantvillage.jsonl'

# Load the datasets
df_gardenology = pd.read_json(gardenology_path, lines=True)
df_plantvillage = pd.read_json(plantvillage_path, lines=True)

# Combine them to check total unique plants
df_combined = pd.concat([df_gardenology, df_plantvillage], ignore_index=True)

print(f"Gardenology rows: {len(df_gardenology)}")
print(f"PlantVillage rows: {len(df_plantvillage)}")
print(f"Total rows: {len(df_combined)}")

if 'plant' in df_combined.columns:
    unique_plants = df_combined['plant'].unique()
    print(f"\nTotal unique plants: {len(unique_plants)}")
    print("\nSample of unique plants:")
    print(unique_plants[:10])
else:
    print("\n'plant' column not found in the combined dataset.")
```
