# Eksplorasi Dataset Korpus, QA, dan Evaluasi
Notebook ini digunakan untuk memperlihatkan sampel data dari korpus dan QA, dengan kolom panjang di-truncate, serta menampilkan hasil evaluasi dari file `evaluation_results.pkl`.

## 1. Import Libraries dan Load Dataset

```python
import pandas as pd
from pathlib import Path

corpus_path = Path('data/corpus.parquet')
qa_path = Path('data/qa.parquet')

if corpus_path.exists():
    df_corpus = pd.read_parquet(corpus_path)
    print('Korpus loaded:', df_corpus.shape)
else:
    df_corpus = None
    print('Korpus file not found')

if qa_path.exists():
    df_qa = pd.read_parquet(qa_path)
    print('QA loaded:', df_qa.shape)
else:
    df_qa = None
    print('QA file not found')
```

## 2. Tampilkan Sampel Data Korpus

```python
if df_corpus is not None:
    sample_corpus = df_corpus.head(5).copy()
    if 'contents' in sample_corpus.columns:
        sample_corpus['contents'] = sample_corpus['contents'].astype(str).str.slice(0, 100) + '...'
    display(sample_corpus)
else:
    print('Korpus belum dimuat.')
```

## 3. Tampilkan Sampel Data QA

```python
if df_qa is not None:
    sample_qa = df_qa.head(5).copy()
    if 'generation_gt' in sample_qa.columns:
        sample_qa['generation_gt'] = sample_qa['generation_gt'].astype(str).str.slice(0, 100) + '...'
    display(sample_qa)
else:
    print('QA belum dimuat.')
```

## 4. Truncate Kolom Panjang pada Sampel

```python
def truncate_string(s, length=100):
    if pd.isnull(s):
        return s
    s = str(s)
    return s[:length] + ('...' if len(s) > length else '')

# Contoh penggunaan pada korpus
df_corpus_trunc = None
df_qa_trunc = None
if df_corpus is not None:
    df_corpus_trunc = df_corpus.copy()
    if 'contents' in df_corpus_trunc.columns:
        df_corpus_trunc['contents'] = df_corpus_trunc['contents'].apply(lambda x: truncate_string(x, 100))
    display(df_corpus_trunc.head(3))

# Contoh penggunaan pada QA
if df_qa is not None:
    df_qa_trunc = df_qa.copy()
    if 'generation_gt' in df_qa_trunc.columns:
        df_qa_trunc['generation_gt'] = df_qa_trunc['generation_gt'].apply(lambda x: truncate_string(x, 100))
    display(df_qa_trunc.head(3))
```

## 5. Load dan Tampilkan Hasil Evaluasi

```python
import os
import pickle

pkl_path = 'evaluation_results.pkl'
if os.path.exists(pkl_path):
    eval_results = pd.read_pickle(pkl_path)
    if isinstance(eval_results, pd.DataFrame):
        display(eval_results.head(5))
    else:
        print(type(eval_results))
        print(str(eval_results)[:500])
else:
    print('File evaluation_results.pkl tidak ditemukan.')
```
